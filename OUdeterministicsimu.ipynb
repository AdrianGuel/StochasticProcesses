```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from scipy import integrate
```

```python
def mu(y, t, gamma,D):
    return -gamma*y

def  beta(y, t, gamma,D):
  return -2*gamma*y+2*D
```

```python
mu0 = 0.25
b0 = 5.0
data=1000
t = np.linspace(0, 10, data)
D=0.01
gamma=2

M = odeint(mu, mu0, t, args=(gamma,D))
plt.plot(t, M, 'g', label=r'$\mu(t)$')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()

Sigma = odeint(beta, mu0, t, args=(gamma,D))
B=1/(2*Sigma)
plt.plot(t, B, 'g', label=r'$\beta(t)$')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
E=((-gamma*M)**2)/Sigma+0.5*((-2*gamma*Sigma+2*D)**2)/(Sigma**2)
plt.plot(t, E, 'g', label=r'$\Gamma^2$')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
E=E.reshape(data)
```

```python
IL = integrate.cumtrapz(np.sqrt(E),t)
fv=len(t)
plt.plot(t[0:fv-1],IL,'g', label=r'$\mathcal{L}$')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
